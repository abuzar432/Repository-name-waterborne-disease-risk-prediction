```python
import pandas as pd
import os

# Step 1: folder path where all CSV files are stored
folder_path = r"C:\AIML waterbourne project\cleaned Lake dataset"

# Step 2: get all CSV files
files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Step 3: create empty list to store dataframes
df_list = []

# Step 4: loop through each file and read
for file in files:
    file_path = os.path.join(folder_path, file)
    
    df = pd.read_csv(file_path)
    
    # OPTIONAL (recommended): add year column
    year = file.split()[0]   # takes 2017 from filename
    df['Year'] = year
    
    df_list.append(df)

# Step 5: merge all datasets
final_df = pd.concat(df_list, ignore_index=True)

# Step 6: save final dataset
final_df.to_csv(r"C:\AIML waterbourne project\final_lake_dataset.csv", index=False)

print("All datasets merged successfully!")
```
