```python
import pandas as pd
final_df = pd.read_csv(r"C:\AIML waterbourne project\final_lake_dataset.csv")
final_df.head()
```

```python
final_df.shape
```

```python
final_df['State Name'] = (
    final_df['State Name']
    .str.replace('\n', ' ', regex=False)
    .str.replace(r'\s+', ' ', regex=True)
    .str.upper()
    .str.strip()
)

final_df['State Name'].unique()
```

```python
san_df = pd.read_csv(r"C:\AIML waterbourne project\final sanitation cleaned_data.csv")
san_df.head()
```

```python
san_df.shape
```

```python
san_df['State'] = (
    san_df['State']
    .str.replace(r'\s+', ' ', regex=True)
    .str.upper()
    .str.strip()
)

san_df['State'].unique()
```

```python
final_df['State Name'] = final_df['State Name'].replace({
    'DELHI': 'NCT OF DELHI',
    'LAKSHADWE EP': 'LAKSHADWEEP',
    'PUDUCHERR Y': 'PUDUCHERRY',
    'CHANDIGAR H': 'CHANDIGARH',
    'CHHATTISGA RH': 'CHHATTISGARH',
    'UTTARAKHA ND': 'UTTARAKHAND'
})

final_df['State Name'].unique()
```

```python
print(final_df.shape)
print(san_df.shape)
```

```python
merged_df = pd.merge(
    final_df,
    san_df,
    left_on=['State Name', 'Year'],
    right_on=['State', 'Year'],
    how='left'
)

merged_df.head()
```

```python
print(merged_df.shape)

print(merged_df.columns)

merged_df.isnull().sum()
```

```python
merged_df['Toilet_Access'] = (
    merged_df.groupby('State Name')['Toilet_Access']
    .transform(lambda x: x.ffill().bfill())
)

merged_df['Open_Defecation'] = (
    merged_df.groupby('State Name')['Open_Defecation']
    .transform(lambda x: x.ffill().bfill())
)

merged_df.isnull().sum()
```

```python
missing_states = merged_df[merged_df['Toilet_Access'].isnull()]['State Name'].unique()

missing_states
```

```python
merged_df['Toilet_Access'] = merged_df['Toilet_Access'].fillna(
    merged_df['Toilet_Access'].mean()
)

merged_df['Open_Defecation'] = merged_df['Open_Defecation'].fillna(
    merged_df['Open_Defecation'].mean()
)

merged_df.isnull().sum()
```

```python
merged_df.drop(columns=['State'], inplace=True)

merged_df.isnull().sum()
```

```python
merged_df.to_csv(
    r"C:\AIML waterbourne project\merged_water_sanitation.csv",
    index=False
)

print("Merged dataset saved successfully!")
```
