```python
import pandas as pd

df = pd.read_csv(
    r"C:\\AIML waterbourne project\\datasets\\cleaned_datasets\\final_complete_dataset.csv"
)

df.head()
```

```python
print(df.columns)
```

```python
def assign_risk(row):

    score = 0

    # Fecal coliform risk
    if row['Min Fecal Coliform'] > 500:
        score += 2
    elif row['Min Fecal Coliform'] > 100:
        score += 1

    # BOD risk
    if row['Min BOD'] > 3:
        score += 2
    elif row['Min BOD'] > 2:
        score += 1

    # Open defecation risk
    if row['Open_Defecation'] > 40:
        score += 2
    elif row['Open_Defecation'] > 20:
        score += 1

    # Toilet access risk
    if row['Toilet_Access'] < 50:
        score += 2
    elif row['Toilet_Access'] < 70:
        score += 1

    # Rainfall risk
    if row['Rainfall'] > 200:
        score += 2
    elif row['Rainfall'] > 100:
        score += 1

    # Temperature risk
    if row['Min Temperature'] > 35:
        score += 2
    elif row['Min Temperature'] > 30:
        score += 1

    # Dissolved Oxygen risk
    if row['Min Dissolved Oxygen'] < 3:
        score += 2
    elif row['Min Dissolved Oxygen'] < 5:
        score += 1

    # pH risk
    if row['Min pH'] < 6 or row['Min pH'] > 8.5:
        score += 2
    elif row['Min pH'] < 6.5 or row['Min pH'] > 8:
        score += 1

    # Final risk level
    if score >= 7:
        return "HIGH"

    elif score >= 4:
        return "MEDIUM"

    else:
        return "LOW"
```

```python
numeric_columns = [
    'Min Fecal Coliform',
    'Min BOD',
    'Open_Defecation',
    'Toilet_Access',
    'Rainfall'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
numeric_cols = [
    'Min Temperature',
    'Min Dissolved Oxygen',
    'Min pH',
    'Min BOD',
    'Min Fecal Coliform',
    'Toilet_Access',
    'Open_Defecation',
    'Rainfall'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Risk_Level'] = df.apply(assign_risk, axis=1)

df.head()
```
