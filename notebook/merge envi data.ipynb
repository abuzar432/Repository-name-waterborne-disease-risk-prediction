```python
import pandas as pd
env_df = pd.read_csv(
    r"C:\AIML waterbourne project\clean environment_data.csv")
env_df.head()
```

```python
print(env_df.shape)

print(env_df.columns)

env_df.isnull().sum()
```

```python
env_df['State'] = (
    env_df['State']
    .str.replace(r'\s+', ' ', regex=True)
    .str.upper()
    .str.strip()
)

env_df['State'].unique()
```

```python
main_df = pd.read_csv(
    r"C:\AIML waterbourne project\merged_water_sanitation.csv"
)

main_df.head()
```

```python
main_df['State Name'].unique()
```

```python
final_merged_df = pd.merge(
    main_df,
    env_df,
    left_on=['State Name', 'Year'],
    right_on=['State', 'Year'],
    how='left'
)

final_merged_df.head()
```

```python
print(final_merged_df.shape)

final_merged_df.isnull().sum()
```

```python
final_merged_df['Rainfall'] = (
    final_merged_df.groupby('State Name')['Rainfall']
    .transform(lambda x: x.ffill().bfill())
)

final_merged_df.isnull().sum()
```

```python
missing_rainfall_states = final_merged_df[
    final_merged_df['Rainfall'].isnull()
]['State Name'].unique()

missing_rainfall_states
```

```python
final_merged_df['Rainfall'] = final_merged_df['Rainfall'].fillna(
    final_merged_df['Rainfall'].mean()
)

final_merged_df.isnull().sum()
```

```python
final_merged_df.drop(columns=['State'], inplace=True)

final_merged_df.isnull().sum()
```

```python
final_merged_df.to_csv(
    r"C:\AIML waterbourne project\final_complete_dataset.csv",
    index=False
)

print("Final dataset saved successfully!")
```
