```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\Mohammed Abuzar\OneDrive\Pictures\Documents\AIML datasets\Indian Rainfall Dataset District-wise Daily Measurements.csv")
df.head()
```

```python
print(df.columns)
```

```python
df.drop(['district', 'month'], axis=1, inplace=True)

df.head()
```

```python
daily_cols = df.columns[1:]  # skip 'state'

df[daily_cols] = df[daily_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df['Rainfall'] = df[daily_cols].sum(axis=1)

df.head()
```

```python
df = df.groupby('state')['Rainfall'].mean().reset_index()
```

```python
df.rename(columns={'state': 'State'}, inplace=True)

df.head()
```

```python
df['Year'] = 2019
```

```python
df['State'] = df['State'].str.upper()
df['State'] = df['State'].str.strip()
```
